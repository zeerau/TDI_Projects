```python
import json

with open('/content/employees_data.json', 'r') as f:
  data = json.load(f)

# 1. Unique departments
departments = set()
for employee in data:
  departments.add(employee['department'])
num_departments = len(departments)
print(f"Number of unique departments: {num_departments}")

# 2. Top 5 highest-paid employees
sorted_employees = sorted(data, key=lambda x: x['salary'], reverse=True)
top_5_employees = sorted_employees[:5]
print(f"Top 5 highest-paid employees: {top_5_employees}")

# 3. Average monthly salary
total_salary = sum(int(employee['salary'].replace('N ', '')) for employee in data)
average_salary = total_salary / len(data)
print(f"Average monthly salary: {average_salary}")

# 4. Employees per department
department_counts = {}
for employee in data:
  department = employee['department']
  department_counts[department] = department_counts.get(department, 0) + 1
print(f"Employees per department: {department_counts}")

# 5. Employees ranked by salary within each department
departments = {}
for employee in data:
  department = employee['department']
  if department not in departments:
    departments[department] = []
  departments[department].append(employee)

for department, employees in departments.items():
  sorted_employees = sorted(employees, key=lambda x: x['salary'], reverse=True)
  print(f"Employees in {department} ranked by salary:")
  for employee in sorted_employees:
    if 'name' in employee:
      print(f"{employee['name']} - {employee['salary']}")
    else:
      print(f"Employee with ID {employee.get('empID', 'Unknown')} - {employee['salary']}") # Print employee ID if name is not available


# 6. Total monthly payroll expense
total_payroll = sum(int(employee['salary'].replace('N ', '')) for employee in data)
print(f"Total monthly payroll expense: {total_payroll}")

# 7. Highest and lowest salary
# highest_salary = max(data, key=lambda x: x['salary'])
# lowest_salary = min(data, key=lambda x: x['salary'])
# # Check if 'name' key exists before accessing it
# print(f"Highest salary: {highest_salary.get('name', highest_salary.get('fullName', f'Employee with ID {highest_salary.get("empID", "Unknown")}'))} - {highest_salary['salary']}")
# print(f"Lowest salary: {lowest_salary.get('name', lowest_salary.get('fullName', f'Employee with ID {lowest_salary.get("empID", "Unknown")}'))} - {lowest_salary['salary']}")
# 8. Remove Junior employees and count remaining
# Add code to remove Junior employees and count remaining employees here
data = [employee for employee in data if employee['experienceLevel'] != 'Junior']
remaining_employees = len(data)
print(f"Number of employees after removing Juniors: {remaining_employees}")
```
